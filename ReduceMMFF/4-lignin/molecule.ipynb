```python
from conformer_rl.molecule_generation.generation.generate_lignin import generate_lignin
from conformer_rl.utils import calculate_normalizers
from rdkit.Chem import AllChem as Chem
from conformer_rl.analysis import drawConformer
```

```python
mol = generate_lignin(4)
mol = Chem.AddHs(mol)
Chem.MMFFSanitizeMolecule(mol)

# E0, Z0 = calculate_normalizers(mol, 500)
```

```python
Chem.EmbedMolecule(mol)
Chem.MMFFOptimizeMolecule(mol, maxIters=2500)
drawConformer(mol)
```
