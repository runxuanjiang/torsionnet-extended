```python
import matplotlib.pyplot as plt
import numpy as np


x = [0, 25, 50, 100, 250, 500]
y_t = [1e-32, 0.1124, 1.242, 1.467, 1.3556, 1.533]
y_e = [0.1221, 0.1382, 1.329, 1.455, 1.492, 1.358]


fig = plt.figure()
ax1 = fig.add_subplot()
ax1.set_xlabel('Max MMFF Iterations (for training only)')
ax1.set_ylabel('Gibbs Score')
ax1.set_title('Gibbs score for alkane over varied max MMFF iters for training, (eval mmff iters always set at 1000)')
ax1.scatter(x, y_t, label='Train')
ax1.scatter(x, y_e, label='Eval')
# ax1.set_yscale('log')
ax1.legend()
```
