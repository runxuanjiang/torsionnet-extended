```python
from rdkit.Chem import Draw
from IPython.display import display
def mol_with_atom_index(mol):
    """call on an rdkit molecule before visualizing to show atom indices in visualization"""
    for atom in mol.GetAtoms():
        atom.SetAtomMapNum(atom.GetIdx())
    return mol
display(Draw.MolToImage(mol_with_atom_index(innerMol.to_rdkit_mol()), size=(700, 300)))
```

```python
for bb in mol.stk_molecule.get_building_blocks():
    display(Draw.MolToImage(mol_with_atom_index(bb.to_rdkit_mol()), size=(700, 300)))
```

```python
def get_torsion_building_block_atoms(torsion):
    building_block_atoms = []
    for atom_info in innerMol.get_atom_infos(torsion):
        building_block_atoms.append(atom_info.get_building_block_atom().get_id())
    return building_block_atoms

print(get_torsion_building_block_atoms([24, 23, 28, 27]))
```
